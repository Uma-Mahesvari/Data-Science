#### **Enhancing Medical Image Quality using Histogram Equalization**

In the realm of digital imaging, the quality of an image plays a crucial role in its usability and visual appeal. Often, images suffer from poor contrast and lighting, making them less effective in conveying information or evoking emotions. Here, histogram equalization can significantly enhance the quality of digital images by improving their contrast and brightness.



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/liver.png', cv2.IMREAD_GRAYSCALE)

# Apply median filter
median_filtered_image = cv2.medianBlur(image, 5)

# Apply histogram equalization
equalized_image = cv2.equalizeHist(median_filtered_image)
```

```python
# Code snippet for visual comparison
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')

plt.subplot(1, 2, 2)
plt.imshow(equalized_image, cmap='gray')
plt.title('Equalized Image')

plt.show()
```

**Histogram Analysis:**

Histograms provide valuable insights into the distribution of pixel intensities within an image. Let's analyze the histograms of the original and equalized images:

```python
# Code snippet for histogram analysis
histogram = cv2.calcHist([image], [0], None, [256], [0, 256])

plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.plot(histogram, color='blue')
plt.plot(122, histogram[122], marker='^', markersize=7, color='none',markeredgecolor="lime",
         markeredgewidth=1)
plt.plot(145, histogram[145], marker='^', markersize=7, color='none',markeredgecolor="red",
         markeredgewidth=1)
plt.title('Original Image Histogram')

plt.subplot(1, 2, 2)
plt.hist(equalized_image.ravel(), 256, [0, 256], color='blue')
plt.title('Equalized Image Histogram')

plt.show()
```
